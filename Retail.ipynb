### 2(c)  “Five-number summary” for continuous variables (min, Q1, median, Q3 and max)


```python
def func_sum(x):
    return pd.Series([x.min() , x.quantile(0.25), x.quantile(0.5), x.quantile(0.75), x.max()] , 
                     index = {'Minimum' ,'Q1' ,'Median','Q3','Maximum'})
```

```python
Customer_Final.select_dtypes(exclude=['object']).apply(func_sum)
```

# 11. For all customers aged between 25 - 35, find out:

### a. What was the total amount spent for “Electronics” and “Books” product categories?

```python
Customer_Final['DOB'] = pd.to_datetime(Customer_Final['DOB'])
```

```python
Customer_Final['Age'] = (pd.to_datetime('today').year - Customer_Final['DOB'].dt.year)
```

```python
Total_cus1 = Customer_Final.loc[((Customer_Final.Age > 25)&(Customer_Final.Age < 35))][['total_amt' , 'prod_cat']]
Total_cus1[((Total_cus1.prod_cat == 'Electronics')|(Total_cus1.prod_cat == 'Books'))].groupby('prod_cat')['total_amt'].sum()
```

```python
print("Total amount spent on 'Electronics' and 'Books' product categories is", 
      Total_cus1.loc[((Total_cus1.prod_cat == 'Electronics')|(Total_cus1.prod_cat == 'Books'))].groupby('prod_cat')['total_amt'].sum().sum())
```

### b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?


```python
Customer_Final['tran_date'] = pd.to_datetime(Customer_Final['tran_date'])
```

```python
start_date = '01-01-2014'
end_date = '01-03-2014'
```

```python
Total_cus2 = Customer_Final.loc[((Customer_Final.Age > 25)&(Customer_Final.Age < 35))][['total_amt' , 'prod_cat' , 'tran_date']]
print('The total amount spent by customers aged 25-35 between 1st Jan 2014 to 1st Mar 2014 is',
      Total_cus2[((Total_cus2.tran_date > start_date)&(Total_cus2.tran_date < end_date))]['total_amt'].sum())
```
